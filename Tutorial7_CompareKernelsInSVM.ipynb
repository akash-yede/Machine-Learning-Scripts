### The eval performance

```python
def eval_performance(classifier,X_train_std,X_test_std,y_train,y_test):
    
    y_pred = classifier.predict(X_test_std)
    
    from sklearn.metrics import accuracy_score
    print("Accuracy of test data set:%.2f"%(accuracy_score(y_test, y_pred)*100)+"%")
    print('Number in test data set: ',len(y_test))
    print('Misclassified samples: %d' % (y_test != y_pred).sum())
    
    X_combined_std = np.vstack((X_train_std, X_test_std))
    y_combined = np.hstack((y_train, y_test))
    
    y_predict_combined=classifier.predict(X_combined_std)
    print('Combined Accuracy: %.2f' %(accuracy_score(y_combined, y_predict_combined)*100)+"%")
    
    print('Number in test ',len(y_combined))
    print('Misclassified samples: %d' % (y_combined != y_predict_combined).sum())
```

```python
b = dft['heart_disease']
```

```python
a = dft.drop(['heart_disease'], axis = 1)
```

### Setting Training and Testing data

```python
from sklearn.model_selection import train_test_split
f_train, f_test, t_train, t_test = train_test_split(a,b,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(f_train)
f_train_std = sc.transform(f_train)
f_test_std = sc.transform(f_test)
```

```python
from sklearn.svm import SVC
SVC_scores=[]
```

### Linear SVM
    Linear Kernel is used when the data is Linearly separable, that is, it can be separated using a single Line. It is one of the most common kernels to be used.
    

```python
SVM_lin = SVC(kernel='linear', C=1.0, random_state=0)
SVM_lin.fit(f_train_std, t_train)
print("\nFor Support Vector Machine(linear):")
eval_performance(SVM_lin,f_train_std,f_test_std,t_train,t_test)
SVC_scores.append(SVM_lin.score(f_test_std,t_test))
```

### Polynomial SVM
    In machine learning, the polynomial kernel is a kernel function commonly used with support vector machines (SVMs) and other kernelized models, that represents the similarity of vectors (training samples) in a feature space over polynomials of the original variables, allowing learning of non-linear models.


```python
SVM_poly = SVC(kernel='poly', C=1.0, random_state=0)
SVM_poly.fit(f_train_std, t_train)
print("\nFor Support Vector Machine(polynomial):")
eval_performance(SVM_poly,f_train_std,f_test_std,t_train,t_test)
SVC_scores.append(SVM_poly.score(f_test_std,t_test))
```

### RBF SVM
    In machine learning, the radial basis function kernel, or RBF kernel, is a popular kernel function used in various kernelized learning algorithms.


```python
SVM_rbf = SVC(kernel='rbf', C=1.0, random_state=0)
SVM_rbf.fit(f_train_std, t_train)
print("\nFor Support Vector Machine(RBF):")
eval_performance(SVM_rbf,f_train_std,f_test_std,t_train,t_test)
SVC_scores.append(SVM_rbf.score(f_test_std,t_test))
```
